```python
#Función para determinar y graficar las variables independientes mas importantes
def plot_feature_importances(clf, X_train, y_train=None, top_n=10, figsize=(8,8), print_table=False, title="Feature Importances"):
    clf.fit(X_train, y_train)
    feat_imp = pd.DataFrame({'importance': clf.feature_importances_})    
    feat_imp['feature'] = X_train.columns
    feat_imp.sort_values(by='importance', ascending=False, inplace=True)
    feat_imp = feat_imp.iloc[:top_n]
    
    feat_imp.sort_values(by='importance', inplace=True)
    feat_imp = feat_imp.set_index('feature', drop=True)
    feat_imp.plot.barh(title=title, figsize=figsize)
    plt.xlabel('Feature Importance Score')
    plt.show()
    
    if print_table:
        from IPython.display import display
        print("Top {} features in descending order of importance".format(top_n))
        display(feat_imp.sort_values(by='importance', ascending=False))
        
    return feat_imp
```

### Stacking

```python
%%time
list_stacking = []
for n, df in enumerate(list_df):
    X_train, X_test, y_train, y_test = train_test_split(df[1].drop(columns='quality'), df[1]['quality'], test_size=0.3, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
    
    clf1 = RandomizedSearchCV(DecisionTreeClassifier(class_weight='balanced'),
                               param_distributions={'min_samples_leaf': randint(5, 101), 'max_depth': randint(3, 10)},
                               n_iter=100, scoring='f1_macro', cv=5)
    clf1.fit(X_train, y_train)
    
    lr_pipe = Pipeline([
                    ('scaler', StandardScaler()),
                    ('log_reg', LogisticRegression(class_weight='balanced', 
                                                multi_class='multinomial',
                                                solver='newton-cg',
                                                fit_intercept=True, 
                                                max_iter=500))])
    dists = {'log_reg__C': reciprocal(1e-4, 1e4)}
    clf2 = RandomizedSearchCV(lr_pipe, param_distributions= dists, n_iter=100, scoring='f1_macro')
    clf2.fit(X_train, y_train)
    
    clf3 = RandomizedSearchCV(RandomForestClassifier(class_weight='balanced'),
                               param_distributions={'n_estimators': randint(10, 101), 'max_depth': randint(3, 10)},
                               n_iter=100, scoring='f1_macro', cv=5)
    clf3.fit(X_train, y_train)
    
    X_new_train = pd.DataFrame ( {
            'Model1': clf1.predict_proba(X_train)[:, 0],
            'Model2': clf2.predict_proba(X_train)[:, 0],
            'Model3': clf3.predict_proba(X_train)[:, 0]
    })
    X_new_val = pd.DataFrame ( {
        'Model1': clf1.predict_proba(X_val)[:, 0],
        'Model2': clf2.predict_proba(X_val)[:, 0],
        'Model3': clf3.predict_proba(X_val)[:, 0]
    })
    clf4 = SVC(kernel='rbf').fit(X_new_train, y_train)
    
    f1_train_1 = f1_score(y_train, clf1.predict(X_train), average='macro')
    f1_val_1 = f1_score(y_val, clf1.predict(X_val), average='macro')

    f1_train_2 = f1_score(y_train, clf2.predict(X_train), average='macro')
    f1_val_2 = f1_score(y_val, clf2.predict(X_val), average='macro')

    f1_train_3 = f1_score(y_train, clf3.predict(X_train), average='macro')
    f1_val_3 = f1_score(y_val, clf3.predict(X_val), average='macro')

    f1_train_4 = f1_score(y_train, clf4.predict(X_new_train), average='macro')
    f1_val_4 = f1_score(y_val, clf4.predict(X_new_val), average='macro')

    print('scores', df[0])
    print('f1 score train clsf1:', f1_train_1)
    print('f1 score val clsf1:', f1_val_1)
    print('--------------------')
    print('f1 score train clsf2:', f1_train_2)
    print('f1 score val clsf2', f1_val_2)
    print('--------------------')
    print('f1 score train clsf3:', f1_train_3)
    print('f1 score val clsf3', f1_val_3)
    print('--------------------')
    print('f1 score train clsf4 (Metamodelo):', f1_train_4)
    print('f1 score val clsf4 (Metamodelo)', f1_val_4)
    print('Classifiction report - val', classification_report(y_val, clf4.predict(X_new_val)))
```

### Stacking (SVM - Random_Forest - Log_Reg)

```python
%%time
list_stacking = []
for n, df in enumerate(list_df):
    X_train, X_test, y_train, y_test = train_test_split(df[1].drop(columns='quality'), df[1]['quality'], test_size=0.3, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
    
    svm_pipe = Pipeline([
                    ('scaler', StandardScaler()),
                    ('svm', SVC(probability=True, kernel='rbf', class_weight='balanced'))])

    dists = {'svm__gamma': reciprocal(0.01, 500), 'svm__C': reciprocal(1e-4, 1e4)}
    clf1 = RandomizedSearchCV(svm_pipe, param_distributions=dists, cv=5, n_iter=100, scoring='f1_macro')
    
    clf1.fit(X_train, y_train)
    
    lr_pipe = Pipeline([
                    ('scaler', StandardScaler()),
                    ('log_reg', LogisticRegression(class_weight='balanced', 
                                                multi_class='multinomial',
                                                solver='newton-cg',
                                                fit_intercept=True, 
                                                max_iter=500))])
    dists = {'log_reg__C': reciprocal(1e-4, 1e4)}
    clf2 = RandomizedSearchCV(lr_pipe, param_distributions= dists, n_iter=100, scoring='f1_macro')
    clf2.fit(X_train, y_train)
    
    clf3 = RandomizedSearchCV(RandomForestClassifier(class_weight='balanced'),
                               param_distributions={'n_estimators': randint(10, 101), 'max_depth': randint(3, 10)},
                               n_iter=100, scoring='f1_macro', cv=5)
    clf3.fit(X_train, y_train)
    
    X_new_train = pd.DataFrame ( {
            'Model1': clf1.predict_proba(X_train)[:, 0],
            'Model2': clf2.predict_proba(X_train)[:, 0],
            'Model3': clf3.predict_proba(X_train)[:, 0]
    })
    X_new_val = pd.DataFrame ( {
        'Model1': clf1.predict_proba(X_val)[:, 0],
        'Model2': clf2.predict_proba(X_val)[:, 0],
        'Model3': clf3.predict_proba(X_val)[:, 0]
    })
    clf4 = SVC(kernel='rbf').fit(X_new_train, y_train)
    
    f1_train_1 = f1_score(y_train, clf1.predict(X_train), average='macro')
    f1_val_1 = f1_score(y_val, clf1.predict(X_val), average='macro')

    f1_train_2 = f1_score(y_train, clf2.predict(X_train), average='macro')
    f1_val_2 = f1_score(y_val, clf2.predict(X_val), average='macro')

    f1_train_3 = f1_score(y_train, clf3.predict(X_train), average='macro')
    f1_val_3 = f1_score(y_val, clf3.predict(X_val), average='macro')

    f1_train_4 = f1_score(y_train, clf4.predict(X_new_train), average='macro')
    f1_val_4 = f1_score(y_val, clf4.predict(X_new_val), average='macro')

    print('scores', df[0])
    print('f1 score train clsf1 - svm:', f1_train_1)
    print('f1 score val clsf1 - svm:', f1_val_1)
    print('--------------------')
    print('f1 score train clsf2 - log_reg:', f1_train_2)
    print('f1 score val clsf2 - log_reg:', f1_val_2)
    print('--------------------')
    print('f1 score train clsf3 - forest:', f1_train_3)
    print('f1 score val clsf3 - forest:', f1_val_3)
    print('--------------------')
    print('f1 score train clsf4 (Metamodelo):', f1_train_4)
    print('f1 score val clsf4 (Metamodelo)', f1_val_4)
    print('Classifiction report - val', classification_report(y_val, clf4.predict(X_new_val)))
```
